```python
# Import packages here
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors
import cartopy.feature as cfeature
import cartopy.crs as ccrs
```

```python
#Reading data
#data source https://www.psehealthyenergy.org/our-work/interactive-tools/california-wildfire/#exceptional
fire = np.genfromtxt("C:/Users/DELL/Desktop/fire/fire.csv",delimiter=",")
lat = fire[:,0]
lon = fire[:,1]
year = fire[:,2]
month = fire[:,3]
size = fire[:,4]
cause = fire[:,5]
```

```python
#Output of the location of fires caused by lightning
cause_light= [] 
for i in range(len(cause)):
    if cause[i] == 1:
        cause_light.append(i)
print(np.shape(cause_light))
lon_light=[]
for j in cause_light:
    lon_light.append(lon[j])
print(np.shape(lon_light))
lat_light=[]
for j in cause_light:
    lat_light.append(lat[j])
print(np.shape(lat_light))

#Output of the location of fires caused by human
cause_human= [] 
for i in range(len(cause)):
    if cause[i] == 2 or cause[i] == 3 or cause[1] == 4 or cause[i] == 6 or cause[i] == 7 or cause[i] == 8 or cause[i] == 10 or cause[i] == 11 or cause[i] == 12 or cause[i] == 13 or cause[i] == 16 :
        cause_human.append(i)
print(np.shape(cause_human))

lon_human=[]
for j in cause_human:
    lon_human.append(lon[j])
print(np.shape(lon_human))
lat_human=[]
for j in cause_human:
    lat_human.append(lat[j])
print(np.shape(lat_human))
```

```python
# Initiate new Figure and Axes  
fig = plt.figure(figsize=(10, 10))
proj = ccrs.PlateCarree()
ax1 = fig.add_subplot(1, 1, 1, projection=proj)

# Draw coastlines
ax1.coastlines()

# Set the projection again, else weird things happen 
# [left_longitude, right_longitude, bottom_latitude, top_latitude]
ax1.set_extent([-125, -114, 42.5, 32], crs=proj)

# Add features
ax1.add_feature(cfeature.LAND)
ax1.add_feature(cfeature.OCEAN)
ax1.add_feature(cfeature.BORDERS, linestyle=':')  # Make dotted lines
ax1.add_feature(cfeature.LAKES, alpha=0.5)  # Make transparent
ax1.add_feature(cfeature.RIVERS)
# ax1.add_feature(states,add)
resolution = '50m'
ax1.add_feature(cfeature.NaturalEarthFeature( 'cultural', 'admin_1_states_provinces_lines', resolution, edgecolor='gray', facecolor='none'))

# Add latitude and longitude values
ax1.set_xticks(range(-125,-113,1), crs=proj)
ax1.set_yticks(range(31,43,1), crs=proj)
ax1.set_xlabel('longitude',fontsize=15)
ax1.set_ylabel('latitude',fontsize=15)
              
              
# Add a title - just as in matplotlib
ax1.set_title('(b) Location map of fires in California from 1980 to 2000', fontsize=15)

ax1.scatter(lon_light, lat_light, c='lightcoral', s=20, marker ='o',label='Lightning')
ax1.scatter(lon_human, lat_human, c='slategray', s=20, marker ='o',label='Human')

# Legend
ax1.legend()
plt.show()
```

```python
#Fires caused by lightning from 1970 to 1978
light_1970_1978=[]
for i in range(len(cause)):
    if cause[i] == 1 and 2001 > year[i] >=1980:
        light_1970_1978.append(i)
print(np.shape(light_1970_1978))

lon_1=[]
for j in light_1970_1978:
    lon_1.append(lon[j])
print(np.shape(lon_1))
lat_1=[]
for j in light_1970_1978:
    lat_1.append(lat[j])
print(np.shape(lat_1))

#Fires caused by humans from 1970 to 1978
human_1970_1978=[]
for i in cause_human:
    if 2001 > year[i] >=1980:
        human_1970_1978.append(i)
print(np.shape(human_1970_1978))

lon_2=[]
for j in human_1970_1978:
    lon_2.append(lon[j])
print(np.shape(lon_2))
lat_2=[]
for j in human_1970_1978:
    lat_2.append(lat[j])
print(np.shape(lat_2))
```

```python
#Fires caused by lightning from 2010 to 2018
light_2010_2018=[]
for i in range(len(cause)):
    if cause[i] == 1 and 2022 > year[i] >=2000:
        light_2010_2018.append(i)
print(np.shape(light_2010_2018))

lon_3=[]
for j in light_2010_2018:
    lon_3.append(lon[j])
print(np.shape(lon_3))
lat_3=[]
for j in light_2010_2018:
    lat_3.append(lat[j])
print(np.shape(lat_3))


#Fires caused by humans from 2010 to 2018
human_2010_2018=[]
for i in cause_human:
    if 2022 > year[i] >=2000:
        human_2010_2018.append(i)
print(np.shape(human_2010_2018))

lon_4=[]
for j in human_2010_2018:
    lon_4.append(lon[j])
print(np.shape(lon_4))
lat_4=[]
for j in human_2010_2018:
    lat_4.append(lat[j])
print(np.shape(lat_4))
```

```python
# Initiate new Figure and Axes  
fig = plt.figure(figsize=(20, 10))
proj = ccrs.PlateCarree()
ax1 = fig.add_subplot(1, 2, 1, projection=proj)
ax2 = fig.add_subplot(1, 2, 2, projection=proj)
# Draw coastlines
ax1.coastlines()
ax2.coastlines()
# Set the projection again, else weird things happen 
# [left_longitude, right_longitude, bottom_latitude, top_latitude]
ax1.set_extent([-125, -114, 42.5, 32], crs=proj)
ax2.set_extent([-125, -114, 42.5, 32], crs=proj)

# Add features
ax1.add_feature(cfeature.LAND)
ax1.add_feature(cfeature.OCEAN)
ax1.add_feature(cfeature.BORDERS, linestyle=':')  # Make dotted lines
ax1.add_feature(cfeature.LAKES, alpha=0.5)  # Make transparent
ax2.add_feature(cfeature.RIVERS)
ax2.add_feature(cfeature.LAND)
ax2.add_feature(cfeature.OCEAN)
ax2.add_feature(cfeature.BORDERS, linestyle=':')  # Make dotted lines
ax2.add_feature(cfeature.LAKES, alpha=0.5)  # Make transparent
ax2.add_feature(cfeature.RIVERS)
# ax1.add_feature(states,add)
resolution = '50m'
ax1.add_feature(cfeature.NaturalEarthFeature( 'cultural', 'admin_1_states_provinces_lines', resolution, edgecolor='gray', facecolor='none'))
ax2.add_feature(cfeature.NaturalEarthFeature( 'cultural', 'admin_1_states_provinces_lines', resolution, edgecolor='gray', facecolor='none'))

# Add latitude and longitude values
ax1.set_xticks(range(-125,-113,1), crs=proj)
ax1.set_yticks(range(31,43,1), crs=proj)
ax1.set_xlabel('longitude',fontsize=15)
ax1.set_ylabel('latitude',fontsize=15)
ax2.set_xticks(range(-125,-113,1), crs=proj)
ax2.set_yticks(range(31,43,1), crs=proj)
ax2.set_xlabel('longitude',fontsize=15)
ax2.set_ylabel('latitude',fontsize=15)              
              
# Add a title 
ax1.set_title('Location map of Fires in California from 1970 to 1978', fontsize=15)
ax2.set_title('Location map of Fires in California from 2010 to 2018', fontsize=15)

ax1.scatter(lon_1, lat_1, c='lightcoral', s=20, marker ='o',label='Lightning')
ax1.scatter(lon_2, lat_2, c='slategray', s=20, marker ='o',label='Human')

ax2.scatter(lon_3, lat_3, c='lightcoral', s=20, marker ='o',label='Lightning')
ax2.scatter(lon_4, lat_4, c='slategray', s=20, marker ='o',label='Human')
# Legend
ax1.legend()
ax2.legend()
plt.show()
```
