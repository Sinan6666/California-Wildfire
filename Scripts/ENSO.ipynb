```python
mul_ENSO = np.genfromtxt("C:/Users/DELL/Desktop/mul_ENSO.csv",delimiter=",",skip_header=1)
print(mul_ENSO.shape)

ENSO = mul_ENSO[:,1]
size = mul_ENSO[:,8]
area_burned = mul_ENSO[:,7]
number = mul_ENSO[:,4]
number_year = mul_ENSO[:,5]
prcp = mul_ENSO[:,6]
monthly_size=mul_ENSO[:,3]
PDSI= mul_ENSO[:,9]
VPD= mul_ENSO[:,10]
Tmax= mul_ENSO[:,11]
month = np.arange(1980,2022,1/12)
```
