```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fire = np.genfromtxt("C:/Users/20453/Desktop/computer_project/FWI.csv",delimiter=",")
year = fire[:,0]
FWI = fire[:,3]
print(np.shape(year))
```

```python
year=range(1978,2019)
Y=50
# Initiate new Figure
fig = plt.figure(figsize=(15,10))

# Set up axes and select map projection
ax1 = fig.add_subplot(1,1,1)

#ax1.plot

ax1.plot(year, FWI, linewidth=2, color='r', label='Surface temperature')

# Add a legend containing the three labels

# ax2.legend()

# add x and y axis labels
ax1.set_xlabel('Year',fontsize=15)
#ax2.set_ylabel('Number of fires',fontsize=15)
ax1.set_ylabel('Fire Weather Index',fontsize=15)


# add a plot title
ax1.set_title('California Fire Weather Index',fontsize=15)

z = np.polyfit(year,FWI,1)
p = np.poly1d(z)
ax1.plot(year,p(year),color='b',label='Trend')
# the line equation:
print ("y=%.6fx+(%.6f)"%(z[0],z[1]))

ax1.legend()
# Show plot
plt.show()
```
