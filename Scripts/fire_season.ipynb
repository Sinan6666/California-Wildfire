```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fire = np.genfromtxt("C:/Users/20453/Desktop/computer_project/fire.csv",delimiter=",")
lat = fire[:,0]
lon = fire[:,1]
year = fire[:,2]
month = fire[:,3]
size = fire[:,4]
```

```python
#from 1970 to 1974
y70_74= [] 
for i in range(len(year)):
    if 1969 < year[i] < 1975:
        y70_74.append(i)
# print(y70_74)

#from 2000 to 2004
y00_04= [] 
for i in range(len(year)):
    if 1999 < year[i] < 2005:
        y00_04.append(i)
        
#from 2014 to 2018
y14_18= [] 
for i in range(len(year)):
    if 2013 < year[i] < 2019:
        y14_18.append(i)
# print(y14_18)
```

```python
Months=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# Initiate new Figure
fig = plt.figure(figsize=(15,10))

# Set up axes and select map projection
ax1 = fig.add_subplot(1,1,1)
ax2 = ax1.twinx()

#ax1.plot

ax1.bar(Months, area_month1, color='lightgrey', label='Area burned from 2014 to 2018')
ax1.bar(Months, area_month2, color='darkgray', label='Area burned from 2000 to 2004')
ax1.bar(Months, area_month, color='grey', label='Area burned from 1970 to 1974')
ax2.plot(Months, number_of_fires, '-^', linewidth=2, markersize=12, color='orange', label='Number of fires from 1970 to 1974')
ax2.plot(Months, number_of_fires2, '-^', linewidth=2, markersize=12, color='blue', label='Number of fires from 2000 to 2004')
ax2.plot(Months, number_of_fires1, '-^', linewidth=2, markersize=12, color='red', label='Number of fires from 2014 to 2018')
# Add a legend containing the three labels
ax2.legend()
ax1.legend(loc='upper left')


# add x and y axis labels
ax1.set_xlabel('Month',fontsize=15)
ax2.set_ylabel('Number of fires',fontsize=15)
ax1.set_ylabel('Area burned (10$^6$Acres)',fontsize=15)


# add a plot title
ax1.set_title('Wildfire and Prescribed Burn Perimeters for each month',fontsize=15)
ax1.set_xticklabels(['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'],fontsize=12)


# Show plot
plt.show()
```
