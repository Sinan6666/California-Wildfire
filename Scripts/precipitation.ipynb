```python
# Import packages here
import numpy as np
from netCDF4 import Dataset
import matplotlib.pyplot as plt
from matplotlib import colors
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

```python
prep = np.genfromtxt("C:/Users/20453/Desktop/computer_project/prep.csv",delimiter=",")
month = prep[:,1]
p = prep[:,2]
```

```python
#total precipitation in June, July and August
June= [] 
for i in range(len(month)):
    if month[i] == 6:
        June.append(i)
p_June=[]
for j in June:
    p_June.append(p[j])
#print(p_June)

July= [] 
for i in range(len(month)):
    if month[i] == 7:
        July.append(i)
p_July=[]
for j in July:
    p_July.append(p[j])
#print(p_July)

Aug= [] 
for i in range(len(month)):
    if month[i] == 8:
        Aug.append(i)
p_Aug=[]
for j in Aug:
    p_Aug.append(p[j])
#print(t_Aug)

# total precipitation in Summer
a = np.array(p_June)
b = np.array(p_July)
c = np.array(p_Aug)
p_summer=a+b+c

print(p_summer)
```

```python
Year = range(1970, 2019)
# Initiate new Figure
fig = plt.figure(figsize=(15,10))

# Set up axes and select map projection
ax1 = fig.add_subplot(1,1,1)

#ax1.plot

ax1.plot(Year, p_summer, linewidth=2, color='r', label='precipitation')
# Add a legend containing the three labels

# ax2.legend()

# add x and y axis labels
ax1.set_xlabel('Year',fontsize=15)
#ax2.set_ylabel('Number of fires',fontsize=15)
ax1.set_ylabel('Precipitation (mm)',fontsize=15)


# add a plot title
ax1.set_title('California Statewide Summer Precipitation Trends',fontsize=15)

z = np.polyfit(Year,p_summer, 1)
p = np.poly1d(z)
ax1.plot(Year,p(Year),color='b',label='Trend')
# the line equation:
print ("y=%.6fx+(%.6f)"%(z[0],z[1]))

ax1.legend()
# Show plot
plt.show()
```
