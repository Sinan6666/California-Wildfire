```python
# Import packages here
import pandas as pd
import geopandas as gpd
import numpy as np
from netCDF4 import Dataset
import matplotlib.pyplot as plt
from matplotlib import colors
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

```python
#Reading data
#data source https://www.psehealthyenergy.org/our-work/interactive-tools/california-wildfire/#exceptional
temp = np.genfromtxt("C:/Users/20453/Desktop/computer_project/temp.csv",delimiter=",")
month = temp[:,1]
t_mean = temp[:,4]
```

```python
#mean temperature in June, July and August
June= [] 
for i in range(len(month)):
    if month[i] == 6:
        June.append(i)
t_June=[]
for j in June:
    t_June.append(t_mean[j])
#print(t_June)

July= [] 
for i in range(len(month)):
    if month[i] == 7:
        July.append(i)
t_July=[]
for j in July:
    t_July.append(t_mean[j])
#print(t_July)

Aug= [] 
for i in range(len(month)):
    if month[i] == 8:
        Aug.append(i)
t_Aug=[]
for j in Aug:
    t_Aug.append(t_mean[j])
#print(t_Aug)
print(type(t_Aug))
#Mean temperature in Summer
a = np.array(t_June)
b = np.array(t_July)
c = np.array(t_Aug)
t_summer=(a+b+c)/3-273.15

print(t_summer)
```

```python
Year = range(1970, 2019)
# Initiate new Figure
fig = plt.figure(figsize=(15,10))

# Set up axes and select map projection
ax1 = fig.add_subplot(1,1,1)

#ax1.plot

ax1.plot(Year, t_summer, linewidth=2, color='r', label='Surface temperature')
# Add a legend containing the three labels

# ax2.legend()

# add x and y axis labels
ax1.set_xlabel('Year',fontsize=15)
#ax2.set_ylabel('Number of fires',fontsize=15)
ax1.set_ylabel(r'Mean Temperature ($\degree$C)',fontsize=15)


# add a plot title
ax1.set_title('California Statewide Summer Trends',fontsize=15)

z = np.polyfit(Year,t_summer, 1)
p = np.poly1d(z)
ax1.plot(Year,p(Year),color='b',label='Trend')
# the line equation:
print ("y=%.6fx+(%.6f)"%(z[0],z[1]))

ax1.legend()
# Show plot
plt.show()
```

```python
#surface T in Sept, Oct and Nov
Sept= [] 
for i in range(len(month)):
    if month[i] == 9:
        Sept.append(i)
t_Sept=[]
for j in Sept:
    t_Sept.append(t_mean[j])
#print(t_Sept)

Oct= [] 
for i in range(len(month)):
    if month[i] == 10:
        Oct.append(i)
t_Oct=[]
for j in Oct:
    t_Oct.append(t_mean[j])
#print(t_Oct)

Nov= [] 
for i in range(len(month)):
    if month[i] == 11:
        Nov.append(i)
t_Nov=[]
for j in Nov:
    t_Nov.append(t_mean[j])
#print(t_Nov)

# total precipitation in Summer
d = np.array(t_Sept)
e = np.array(t_Oct)
f = np.array(t_Nov)
t_autumn=(d+e+f)/3-273.15

print(t_autumn)
```

```python
Year = range(1970, 2019)
# Initiate new Figure
fig = plt.figure(figsize=(15,10))

# Set up axes and select map projection
ax1 = fig.add_subplot(1,1,1)

#ax1.plot

ax1.plot(Year, t_autumn, linewidth=2, color='r', label='Surface temperature')
# Add a legend containing the three labels

# ax2.legend()

# add x and y axis labels
ax1.set_xlabel('Year',fontsize=15)
#ax2.set_ylabel('Number of fires',fontsize=15)
ax1.set_ylabel(r'Mean temperature ($\degree$C)',fontsize=15)


# add a plot title
ax1.set_title('California Statewide Autumn Temperature Trends',fontsize=15)

z = np.polyfit(Year,t_autumn, 1)
p = np.poly1d(z)
ax1.plot(Year,p(Year),color='b',label='Trend')
# the line equation:
print ("y=%.6fx+(%.6f)"%(z[0],z[1]))

ax1.legend()
# Show plot
plt.show()
```
